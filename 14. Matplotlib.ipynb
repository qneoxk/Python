# Fonksiyon grafikleri

```python
plt.figure(figsize = (18,6))

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1/4)
y2 = np.cos(2 * np.pi * x2)

plt.title("Sönümlü ve sönümsüz salınıcı")

# plot 1:
plt.subplot(2, 1, 1)
plt.scatter(x1, y1, color = "darkblue", marker = "o", s = 20, label = "Sönümlü salınım")
plt.plot(x1, y1, color = "red")
plt.grid()
plt.legend()

# plot 2:
plt.subplot(2, 1, 2)
plt.scatter(x2, y2, color = "black", marker = "o", s = 20, label = "Sönümsüz salınım")
plt.plot(x2, y2, color = "red")
plt.grid()
plt.legend()

plt.show()
```

```python
def sin(x):
    return np.sin(np.deg2rad(x))
```

```python
x = np.arange(0, 365, 5)
```

```python
plt.figure(figsize = (6,3))

plt.title("Sin(x) fonksiyonunun grafiği")

plt.scatter(x, sin(x), color = "darkblue", s = 5, label = "Sin(x)")
plt.plot(x, sin(x), color = "red", label = "Sin(x)")

plt.xlabel("Açı")
plt.ylabel("Sin(x)")

plt.grid()
plt.legend()
plt.show()
```

<span style = "color : red; font-size: 42px;"> 3D grafik çizimi </span>

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.title('Temel 3D Scatter Grafiği')

# Noktaları tanımlayalım
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
z = [5, 4, 3, 2, 1]

# 3D scatter çizimi
ax.scatter(x, y, z, color='r', marker='o')

ax.set_xlabel('X Ekseni')
ax.set_ylabel('Y Ekseni')
ax.set_zlabel('Z Ekseni')

plt.show()
```

<span style = "color : red; font-size: 42px;"> Yüzey grafiği </span>

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Grid oluştur
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)

# Fonksiyon tanımı
z = np.sin(np.sqrt(x**2 + y**2))

# Surface grafiği
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Renk çubuğu (colorbar)
fig.colorbar(surf, ax=ax)

# 🔄 Bakış açısını ayarlayalım
ax.view_init(elev=30, azim=45)  # 30 derece yukarıdan, 45 derece yandan

ax.set_title('3D Yüzey Grafiği')
plt.show()
```

<span style = "color : red; font-size: 42px;"> WireFrame (tel kafes - iskelet) grafiği </span>

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.cos(x) * np.sin(y)

ax.plot_wireframe(x, y, z, color="darkblue")
ax.view_init(elev=60, azim=45)  # 30 derece yukarıdan, 45 derece yandan

ax.set_title('3D Wireframe Grafiği') # Tel kafes / iskelet
plt.show()
```

<span style = "color : red; font-size: 42px;"> 3D vektör alanı </span>

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D figür oluştur
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Nokta ağı (grid)
x, y, z = np.meshgrid(
    np.linspace(-1, 1, 5),
    np.linspace(-1, 1, 5),
    np.linspace(-1, 1, 5)
)

# Vektör bileşenleri (örnek: merkezden dışarı yayılan alan)
u = x
v = y
w = z

# Vektörleri çiz
ax.quiver(x, y, z, u, v, w, length=0.4, normalize=True, color='purple')

# Eksenleri etiketle
ax.set_xlabel('X Ekseni')
ax.set_ylabel('Y Ekseni')
ax.set_zlabel('Z Ekseni')
ax.set_title('3D Vektör Alanı (Merkezden dışarı)')
plt.show()
```

<span style = "color : red; font-size: 42px;"> surface + wireframe </span>

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.cos(x) * np.sin(y)

ax.plot_surface(x, y, z, cmap="coolwarm", alpha=0.6)
ax.plot_wireframe(x, y, z, color="black", linewidth=0.3)

ax.view_init(elev=40, azim=45)  # 30 derece yukarıdan, 45 derece yandan

ax.set_title('3D Wireframe Grafiği') # Tel kafes / iskelet
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Grid (x-y düzlemi)
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)

# Z yüzeyi (yükseklik): dalgalı fonksiyon
z = np.sin(np.sqrt(x**2 + y**2))

# Grafik oluştur
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 🎨 Yüzey grafiği
surf = ax.plot_surface(x, y, z, cmap='viridis', alpha=0.8)

# 🕸️ Wireframe grafiği (aynı grid üzerine)
ax.plot_wireframe(x, y, z, color='black', linewidth=0.5)

# Renk çubuğu
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

# Eksen etiketleri
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Surface + Wireframe 3D Grafiği')

# Bakış açısı
ax.view_init(elev=30, azim=45)  # İsteğe göre değiştir

plt.show()
```
